```python
def solution(clothes):
    sum=0
    from collections import Counter
    from itertools import combinations 
    re=[]
    for i in clothes:
        re.append(i[1])
    if len(set(re))==1:
        answer=len(re)
    else:
        c=Counter(re)
        lis=[]
        for i,j in c.items():
          lis.append(j) 
          answer=sum(lis)
          for i in range(2,len(lis)+1):
            tu=list(combinations(lis, i))
            for i in tu:
              answer+=eval('*'.join([str(n) for n in i]))
    return answer
```

```python
#참고한 답
from itertools import combinations

def solution(clothes):
    answer = 1
    dic = dict()

    for cloth, Type in clothes:
        if(dic.get(Type) == None ):
            dic[Type] = [cloth]
            continue
        dic[Type].append(cloth)    

    for Type in dic:
        answer *= len(dic[Type]) + 1

    return answer - 1
```
